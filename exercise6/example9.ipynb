Task 6.1: Combine year 2019 and year 2020’s crime records into one pandas dataframe. Drop duplicates as well as any record that has missing values in the following columns: “INCIDENT_NUMBER”, “OFFENSE_CODE”, “DISTRICT”, “Long”, “Lat”, “SHOOTING”, “OCCURRED_ON_DATE”. Save the “cleaned” pandas dataframe to a separate csv file. 

```python
import pandas as pd
Boston_crime_2019 = pd.read_csv("/Users/[editted]/Downloads/tmp9mkqyv6b.csv")
Boston_crime_2019.head()
```

```python
Boston_crime_2020 = pd.read_csv("/Users/[editted]/Downloads/script_113631134_20210423193017_combine.csv")
Boston_crime_2020.head()
```

```python
Boston_crime_merged_data = Boston_crime_2019.append(Boston_crime_2020)
Boston_crime_merged_data.head()
```

```python
Boston_crime_merged_data.drop_duplicates()
```

```python
Boston_crime_merged_dropped = Boston_crime_merged_data.dropna(subset=['INCIDENT_NUMBER', 'OFFENSE_CODE', 'DISTRICT', 'Long', 'Lat', 'SHOOTING', 'OCCURRED_ON_DATE'])
Boston_crime_merged_dropped.to_csv("/Users/[editted]/Downloads/boston_crime.csv")
```
