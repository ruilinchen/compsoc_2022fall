```python
#human mobility network
import os
import networkx as nx
import pandas as pd
os.chdir('/Users/[editted]/Desktop/class related docs/compSoc/code/assignment10/MA_cbg_human_mobility_2019')
ma_df = pd.read_csv('MA_cbg2cbg_2019_01.csv')
G = nx.Graph()
for i in range(20):
    origin_cbg = ma_df.loc[i, 'poi_cbg']
    destination_cbg = ma_df.loc[i, 'visitor_census_tract']
    total_residents_in_origin = ma_df.loc[i, 'raw_device_counts']
    num_visitor = ma_df.loc[i, 'visitor_count']
    percent_visitor = num_visitor / total_residents_in_origin
    print(origin_cbg, destination_cbg, percent_visitor)
    G.add_edge(origin_cbg, destination_cbg, weight=percent_visitor)
print(nx.info(G))
```

```python
nx.draw(G, with_labels=True)
```

```python
#node dataframe
ma_df = pd.read_csv('MA_cbg2cbg_2019_01.csv')
id = G.nodes
unweighted = []
weighted = []
for i in range(20):
    unweighted = G.degree(id)
    weighted = G.degree(id, weight='weight')
df = pd.DataFrame()
df['groupID'] = id
df['weighted node degree'] = weighted
df['unweighted node degree'] = unweighted
print(df)
```

```python
#weighted node degrees plot
import seaborn as sns
sns.barplot(x = df['groupID'], y = df['weighted node degree'], data = df)
```
