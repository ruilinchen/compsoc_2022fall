```python
import os
import numpy as np
import pandas as pd
import spacy

nlp = spacy.load('en_core_web_sm')

path = "/Users/[editted]/Desktop/compSoc/datasets/indeed_scraped_data/job_info_data"
os.chdir(path)
fol = os.listdir(path)

li = []

for f in fol:
    if ".csv" in f and "522" in f:
        df = pd.read_csv(f)
        li.append(df)

df = pd.concat(li)
df.dropna(subset = ["lnks_job_title", "lnks_job_description"], inplace = True)

token2tag = {}

for i in range(len(df)):
    job_info = df.iloc[i]['lnks_job_description']
    for token in nlp(job_info):
        token2tag[token.text] = token.pos_

token_df = pd.DataFrame(columns=['token', 'tag'])
token_df['token'] = token2tag.keys()
token_df['tag'] = token2tag.values()

print('unique tokens:', len(token2tag))
print('distribution:')
print(token_df['tag'].value_counts())
```

```python
import os
import numpy as np
import pandas as pd
import spacy
import json

nlp = spacy.load('en_core_web_sm')

path = "/Users/[editted]/Desktop/compSoc/datasets/indeed_scraped_data/job_info_data"
os.chdir(path)
fol = os.listdir(path)

li = []

for f in fol:
    if ".csv" in f and "522" in f:
        df = pd.read_csv(f)
        li.append(df)

df = pd.concat(li)
df.dropna(subset = ["lnks_job_title", "lnks_job_description"], inplace = True)

unique_tokens = set()
all_tokens = []
for i in range(len(df)):
    job_info = df.iloc[i]['lnks_job_description']
    for token in nlp(job_info):
        unique_tokens.add(token.text)
        all_tokens.append(token.text)

token_dict = {}
for i in unique_tokens:
    token_dict[i] = all_tokens.count(i)
    
json_obj = json.dumps(token_dict, indent = 4)

with open("tokens.json", 'w') as outFile:
    outFile.write(json_obj)
    
print(token_dict)
```
