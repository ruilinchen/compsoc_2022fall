```python
#bubble map for property crimes in 2019
import geopandas as gpd
import pandas as pd
import folium
import random
crime_df = pd.read_csv('cleanedcrime.csv')
boston_gdf = gpd.read_file('Boston_Zoning_Subdistricts.geojson')

crime_df['OFFENSE_TYPE'] = None
crime_df['OFFENSE_TYPE'][crime_df['OFFENSE_DESCRIPTION'].str.lower().str.contains('robbery|assault|rape')] = 'violent crime'
crime_df['OFFENSE_TYPE'][crime_df['OFFENSE_DESCRIPTION'].str.lower().str.contains('burglary|larceny|theft|arson')] = 'property crime'
crime_df = crime_df.groupby(['OFFENSE_TYPE', 'YEAR', 'DISTRICT', 'Long', 'Lat']).agg({'INCIDENT_NUMBER': 'count'}).reset_index()

gpd.points_from_xy(crime_df['Long'], crime_df['Lat'])
crime_gdf = gpd.GeoDataFrame(crime_df, geometry=gpd.points_from_xy(crime_df['Long'], crime_df['Lat']))
crime_gdf.set_crs(epsg=4326, inplace=True)

crime_gdf = crime_gdf[crime_gdf['Long'] < 0].reset_index()
clon, clat = crime_gdf['Long'].mean(), crime_gdf['Lat'].mean()
m = folium.Map(location=(clat, clon), zoom_start=12, width=800, height=800)

for i in range(50):
    crime_index = random.randrange(0, len(crime_gdf))
    crime_long, crime_lat = crime_gdf.loc[crime_index, 'Long'], crime_gdf.loc[crime_index, 'Lat']
    offense_type = crime_gdf.loc[crime_index, 'OFFENSE_TYPE']
    year = crime_gdf.loc[crime_index, 'YEAR']
    if offense_type == 'property crime':
        folium.CircleMarker((crime_lat, crime_long), opacity=0.7, color='red', weight=4).add_to(m)
    if year == '2020':
        folium.CircleMarker((crime_lat, crime_long), opacity=0.7, color='clear', weight=4).add_to(m)
    else:
        folium.CircleMarker((crime_lat, crime_long), opacity=0.7, color='clear', weight=4).add_to(m)
display(m)
```
