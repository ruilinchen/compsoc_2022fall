```python
job_title_lower = []
for title in job_title:
    job_title_lower.append(title.lower())
```

```python
token2tag = {}

for i in range(len(job_title)):
    job_title_token = job_title[i]
    for token in nlp(job_title_token):
    # print(token.text, token.pos_, token.dep_)
        token2tag[token.text] = token.pos_

token_df = pd.DataFrame(columns=['token', 'tag'])
token_df['token'] = token2tag.keys()
token_df['tag'] = token2tag.values()

print('number of unique tokens in the job description:', len(token2tag))
print('the distribution of token tags among them:')
print(token_df['tag'].value_counts())
```

```python
for title in job_title:
    doc = nlp(title)
    for chunk in doc.noun_chunks:
        print([
            chunk,            # A Span object with the full phrase.
            chunk.root,       # The key Token within this phrase.
            chunk.root.dep_,  # The grammatical role of this phrase.
            chunk.root.head   # The grammatical parent Token.
        ])
```

```python
noun_title = []
for title in job_title_lower:
    doc = nlp(title)
    for chunk in doc.noun_chunks:
        if chunk.root.head.pos_ == 'NOUN':
        # print([chunk, chunk.root, chunk.root.dep_, chunk.root.head.lemma_])
            noun = (chunk.root.head.lemma_)
            noun_title.append(noun)
            print('noun:', noun)
```

```python
len(noun_title)
```

```python
adj_title = []
for title in job_title_lower:
    doc = nlp(title)
    for chunk in doc.noun_chunks:
        if chunk.root.head.pos_ == 'ADJ':
        # print([chunk, chunk.root, chunk.root.dep_, chunk.root.head.lemma_])
            adj = (chunk.text, chunk.root.head.lemma_)
            adj_title.append(adj)
            print('adj:', adj)
```

```python
len(adj_title)
```

The adjectives give clues about the nature of the job such as remote vs in person. While nouns give us the concept of the jobsuch as specialist, manager etc.

### Task 3. 
Use spacy's word vector to do Task 2. Compare the results. 

```python
# Similarity of two documents

similarity = []
for title in job_title_lower:
    doc1 = nlp(job_title_lower[0]) 
    doc2 = nlp(title)
    similarity.append(
        {"doc1" : doc1, 
        "doc2": doc2, 
        "similarity": doc1.similarity(doc2)
        }
    )
    print(doc1, "<->", doc2, doc1.similarity(doc2))
```

```python
similarity = pd.DataFrame(similarity)
similarity.sort_values(by='similarity', ascending=False)
```

Spacy gives a much more accurate description and shows us a spectrum of similarity.
