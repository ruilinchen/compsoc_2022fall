```python
import pandas as pd
import os
from tqdm import tqdm
import json
import numpy as np
import seaborn as sns
import spacy
nlp = spacy.load('en_core_web_sm')
```

## task13.1

```python
os.chdir('/Users/edu/Documents/compsoc/code/indeed_scraped_data/job_info_data')
```

```python
info_0522_df = pd.concat(map(pd.read_csv, ['parsehub_5222022_v1.csv', 'parsehub_5222022_v2.csv', 'parsehub_5222022_v3.csv', 'parsehub_5222022_v4.csv', 'parsehub_5222022_v5.csv', 'parsehub_5222022_v6.csv']), ignore_index=True)
print(info_0522_df)
```

```python
info_0522_df = info_0522_df.dropna()
info_0522_df
```

```python
all_title = ""
for item in info_0522_df.iloc:
    title = str(item["lnks_job_title"])
    all_title += (". " + title)

tokens = nlp(all_title)
```

```python
noun_token_list = []
adj_token_list = []
for t in tokens:
    if t.pos_ == "NOUN":
        if t.is_alpha:
            noun_token_list.append(t.lemma_.lower())
        else:
            noun_token_list.append(t.text)
    if t.pos_ == "ADJ":
        if t.is_alpha:
            adj_token_list.append(t.lemma_.lower())
        else:
            adj_token_list.append(t.text)
```

```python
# How many unique vocabulary set are there?
print("Noun:", len(set(noun_token_list)))
print("Adj:", len(set(adj_token_list)))
```

With the increase in the number of adjectives, we have reason to believe that this part of the information contains more subjective intentions than a job description consisting mainly of nouns. The increase in the number of nouns will give people a sense of "it is what it is"

## task13.2

```python
vocab_with_freq = dict()

for i in tqdm(range(len(info_0522_df))):
    job_title = info_0522_df.iloc[i]['lnks_job_title']
    for token in nlp(job_title):
        if token.is_alpha:
            vocab_with_freq[token.lemma_.lower()] = vocab_with_freq.get(token.lemma_.lower(), 0) + 1
        else:
            vocab_with_freq[token.text] = vocab_with_freq.get(token.lemma_, 0) + 1

vocabulary_list = list(vocab_with_freq.keys())
print('there are', len(vocabulary_list), 'unique words among indeed job titles')
```

```python
one_hot_encodings = []
noun_token_set = list(set(noun_token_list))
adj_token_set = list(set(adj_token_list))

# for i in tqdm(range(len(job_data_clean))):
#     job_title = job_data_clean.iloc[i]['lnks_job_title'] # take one example first
token_indices = []

for t in tokens:
    if t.pos_ != "NOUN":
        continue
    if t.is_alpha:
        token_index_in_vocab = noun_token_set.index(t.lemma_.lower())
        token_indices.append(token_index_in_vocab)
    else:
        token_index_in_vocab = noun_token_set.index(t.text)
        token_indices.append(token_index_in_vocab)

    one_hot_encoding = np.zeros(len(noun_token_set))
    for token_index in token_indices:
        one_hot_encoding[token_index] = 1

    one_hot_encodings.append(one_hot_encoding)
```

## task13.3

```python
from scipy.spatial.distance import cosine

onecodings_res = []
print("-------------------------cosine similarity with \"", info_0522_df.iloc[0]['lnks_job_title'], "\"-------------------------")

for i in range(1, len(one_hot_encodings)):
    if (1 - cosine(one_hot_encodings[0], one_hot_encodings[i])) > 0:
        line = info_0522_df.iloc[i]['lnks_job_title']
        print(line)
        onecodings_res.append(line)
```

```python
v_encodings = []
for t in tokens:

    v_encodings.append(t.vector)
```

```python
print("-------------------------cosine similarity with ", info_0522_df.iloc[0]['lnks_job_title'], "-------------------------")

vector_res = []
for i in range(1, len(v_encodings)):
    if (1 - cosine(v_encodings[0], v_encodings[i])) > 0:
        line = info_0522_df.iloc[i]['lnks_job_title']
        print(line)
        vector_res.append(line)
```
