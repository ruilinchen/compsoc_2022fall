```python
#task6.1

##concat
import os

os.chdir('E:\jycompsoc\code\crimereport')

import pandas as pd
crime2019_df = pd.read_csv('Bostoncrimereport2019.csv', low_memory=False)
print(crime2019_df)

crime2020_df= pd.read_csv('bostoncrimereport2020.csv', low_memory=False)
print(crime2020_df)

combined_df = pd.concat([crime2019_df, crime2020_df])
print(combined_df)
```

```python
##duplicate
print(combined_df[['INCIDENT_NUMBER', 'OFFENSE_CODE', 'DISTRICT', 'Long', 'Lat','SHOOTING','OCCURRED_ON_DATE']].isna().sum())
dropnullreport_df=combined_df.dropna(subset=['INCIDENT_NUMBER', 'OFFENSE_CODE', 'DISTRICT', 'Long', 'Lat','SHOOTING','OCCURRED_ON_DATE'])
print(dropnullreport_df)
```

```python
##missing values
print(dropnullreport_df.duplicated(subset=['INCIDENT_NUMBER', 'OFFENSE_CODE', 'DISTRICT', 'Long', 'Lat','SHOOTING','OCCURRED_ON_DATE']).sum())
new_df=dropnullreport_df.drop_duplicates(subset=['INCIDENT_NUMBER', 'OFFENSE_CODE', 'DISTRICT', 'Long', 'Lat','SHOOTING','OCCURRED_ON_DATE'])
print(new_df)
```

```python
##csv file
new_df.to_csv('newfile.csv', index=False)
```
