```python
##task3

##What is the percentage of jobs that can be matched between these two data sources?
 

###raw data of jobinfo
match_df= cleaned2_df.merge(cleaned_jobinfo, how='left', left_on='job_url', right_on='lnks_link')


match_df['has_null'] = match_df['lnks_link'].isna()
print(match_df['has_null'].describe())
match_df


######Answer:16303/21260=76.68%
```

```python
##How are the missing data (unmatched job urls) distributed by date?
unmatchws_by_date = match_df.groupby(['date','has_null']).agg({'job_url':'count'}).reset_index()
unmatchws_by_date 

### Answer: unmatched job urls are mainly located on May 17th, 22nd, and 23rd.The majority of job urls of May 18th, 19th, 20th, 21st are matched.
```
