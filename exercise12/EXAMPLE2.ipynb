## Task 1. 
Create one pandas dataframe that combines all the data scraped from May 22, 2022 together. Drop rows with missing job titles and/or job descriptions. Use `spacy` to tokenize all the job descriptions included in the cleaned dataframe. Count how many unique tokens are there in total, and describe the distribution of token tags among them.

```python
import os
import pandas as pd
import spacy
import re
import json
nlp = spacy.load('en_core_web_sm')
os.chdir('/Users/[editted]/Dropbox/work/compsoc/dataset/indeed_scraped_data/job_info_data')
os.listdir()
```

```python
all_df=[]
for item in list(os.listdir()):
    if "5222022" in item:
        df = pd.read_csv(item)
        all_df.append(df)

jobinfo=pd.concat(all_df).dropna(subset=['lnks_job_title', 'lnks_job_description']).reset_index(drop=True)
jobinfo 
```

```python
token2tag = {}
for item in jobinfo['lnks_job_description']:
    for token in nlp(item):
        token2tag[token.text] = token.pos_

token_df = pd.DataFrame(columns=['token', 'tag'])
token_df['token'] = token2tag.keys()
token_df['tag'] = token2tag.values()

print('number of unique tokens in the job description:', len(token2tag))
print('the distribution of token tags among them:')
print(token_df['tag'].value_counts())
```

## Task 2. 
Construct a token/term frequency dictionary from the cleaned dataframe. The keys of the dictionary should be unique tokens, and the values being the frequency of the tokens in the entire corpus. Save that dictionary to a json file. What are the top 10 most common tokens according to your frequency dictionary? 


```python
token_list = []
for item in jobinfo['lnks_job_description']:
    for token in nlp(item):
        token_list.append(token.text)
```

```python
from collections import Counter
print('number of tokens in the job description:', len(token_list))
print(Counter(token_list).most_common(10))
```

```python
tf_dic=dict(Counter(token_list))
with open("/Users/[editted]/Dropbox/work/compsoc/dataset/indeed_scraped_data/token_freq.json", "w") as f:
    json.dump(tf_dic, f)
```
